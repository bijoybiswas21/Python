## What is Matplotlib?

**Matplotlib** is a popular Python library for creating static, interactive, and animated visualizations in Python. It is widely used for plotting graphs, charts, and figures in data analysis and scientific research.

---

## Key Features

- **2D and 3D plotting** (with `mpl_toolkits.mplot3d`)
- Highly customizable plots
- Supports various output formats (PNG, PDF, SVG, EPS, etc.)
- Integrates with NumPy, Pandas, and other libraries
- Interactive plotting in Jupyter Notebooks

---

## Basic Components

- **Figure**: The entire window or page that everything is drawn on.
- **Axes**: The area where data is plotted (can have multiple axes in a figure).
- **Axis**: The x and y (and optionally z) axis in a plot.
- **Artist**: Everything you see on the figure (title, lines, labels, etc.).

---

## Basic Usage

```python
import matplotlib.pyplot as plt

# Simple line plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

---

## Common Plot Types

- **Line Plot**: `plt.plot()`
- **Scatter Plot**: `plt.scatter()`
- **Bar Chart**: `plt.bar()`, `plt.barh()`
- **Histogram**: `plt.hist()`
- **Pie Chart**: `plt.pie()`
- **Box Plot**: `plt.boxplot()`
- **3D Plot**: `Axes3D.plot()`

---

## Customization

- **Titles and Labels**: `plt.title()`, `plt.xlabel()`, `plt.ylabel()`
- **Legends**: `plt.legend()`
- **Grid**: `plt.grid(True)`
- **Colors and Styles**: Use arguments like `color`, `linestyle`, `marker`
- **Subplots**: `plt.subplot()`, `plt.subplots()`

---

## Example: Multiple Plots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [30, 25, 20, 10]

plt.plot(x, y1, label='Ascending')
plt.plot(x, y2, label='Descending')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Multiple Lines')
plt.legend()
plt.show()
```

## Integration with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df.plot(kind='bar')
plt.show()
```

````python
import matplotlib.pyplot as plt
x=['python', 'c', 'c++', 'java', 'javascript']
y=[85, 70, 60, 50, 81]
plt.xlabel('Languages', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.ylabel('Popularity', fontsize=14)
plt.yticks(fontsize=12)
plt.title('Programming Language Popularity', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
c=['purple', 'orange', 'blue', 'green', 'red']
plt.bar(x, y, color=c,edgecolor='black',)
plt.show()
# This code creates a bar chart showing the popularity of different programming languages.
# The x-axis represents the languages, and the y-axis represents their popularity scores.
# The chart includes labels for the axes, a title, and grid lines for better readability.
# The bars are colored purple, and the x-axis labels are rotated for better visibility.
# The chart is displayed using plt.show().
# This code creates a bar chart showing the popularity of different programming languages.
# The x-axis represents the languages, and the y-axis represents their popularity scores.
````
