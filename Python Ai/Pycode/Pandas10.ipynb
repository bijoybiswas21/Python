```python
# Import pandas library
import pandas as pd

# Create a Pandas Series
data = [10, 20, 30, 40, 50]
series = pd.Series(data, index=['A', 'B', 'C', 'D', 'E'])

# Display the Series
print("Original Series:")
print(series)

# Perform operations on the Series
print("\nSeries after adding 5 to each element:")
print(series + 5)

# Accessing elements
print("\nAccessing element at index 'C':")
print(series['C'])
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
}
df = pd.DataFrame(data)

# Plot the data
plt.bar(df['Category'], df['Values'], color='skyblue')

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [250, 300, 400, 350, 500]
}
df = pd.DataFrame(data)

# Plot the data as a line chart
plt.plot(df['Year'], df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.legend()

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
}
df = pd.DataFrame(data)

# Plot the data as a pie chart
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange', 'green', 'red'])

# Add a title
plt.title('Pie Chart Example')

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Values': [10, 20, 20, 30, 30, 30, 40, 40, 50, 50, 50, 50]
}
df = pd.DataFrame(data)

# Plot the data as a histogram
plt.hist(df['Values'], bins=5, color='purple', edgecolor='black')

# Add labels and title
plt.xlabel('Value Ranges')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
x = np.arange(1, 6)
y = np.array([2.5, 3.5, 4.0, 4.5, 5.0])
error = np.array([0.2, 0.3, 0.1, 0.4, 0.2])

# Plot the data with error bars
plt.errorbar(x, y, yerr=error, fmt='o', color='blue', ecolor='red', capsize=5, label='Data with Error')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bars Chart Example')
plt.legend()

# Show the plot
plt.show()
```
